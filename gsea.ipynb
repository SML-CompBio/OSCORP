```python
expr_data = pd.read_csv('22_tcga.csv', index_col=(0))
expr_data = expr_data.sort_index().T
expr_data
```

```python
from adjustText import adjust_text

def plot_mtcars(adjust=False, *args, **kwargs):
    plt.figure(figsize=(9, 6))
    plt.scatter(gs_res['index'], gs_res['NES'], s=15, c='r', edgecolors=(1,1,1,0))
    texts = []
    for x, y, s in zip(gs_res['index'], gs_res['NES'], gs_res['Term']):
        texts.append(plt.text(x, y, s, size=5))
    plt.xlabel('')
    plt.ylabel('NES')
    if adjust:
        adjust_text(
            texts,
            arrowprops=dict(arrowstyle="-", color='k', lw=0.5),
            text_from='points',
            expand_points=(1.5, 1.5),
        )
plot_mtcars()
```

```python
import networkx as nx
```

```python
from gseapy import enrichment_map
nodes, edges = enrichment_map(gs_res_hesc.res2d)
```

```python
G = nx.from_pandas_edgelist(edges,
                            source='src_idx',
                            target='targ_idx',
                            edge_attr=['jaccard_coef', 'overlap_coef', 'overlap_genes'])
```

```python
print(G)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

# init node cooridnates
pos=nx.layout.spiral_layout(G)
#node_size = nx.get_node_attributes()
# draw node
nx.draw_networkx_nodes(G,
                       pos=pos,
                       cmap=plt.cm.RdYlBu,
                       node_color=list(nodes.NES),
                       node_size=list(nodes.Hits_ratio *1000))
# draw node label
nx.draw_networkx_labels(G,
                        pos=pos,
                        labels=nodes.Term.to_dict())
# draw edge
edge_weight = nx.get_edge_attributes(G, 'jaccard_coef').values()
nx.draw_networkx_edges(G,
                       pos=pos,
                       width=list(map(lambda x: x*10, edge_weight)),
                       edge_color='#CDDBD4')

# plt.savefig('network_hmsc_husc.png', dpi=600, bbox_inches='tight')
plt.show()
```
